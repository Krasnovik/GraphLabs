```python
from queue import PriorityQueue

def bellman_ford(graph, source):
    distances = initialize_distances(graph, source)

    # Relax edges V-1 times
    for _ in range(len(graph) - 1):
        for from_vertex in range(len(graph)):
            for to_vertex in range(len(graph[from_vertex])):
                weight = graph[from_vertex][to_vertex]
                if weight != 0:  # Игнорируем нулевые веса (отсутствие ребра)
                    relax(distances, from_vertex, to_vertex, weight)

    # Check for negative cycles
    for from_vertex in range(len(graph)):
        for to_vertex in range(len(graph[from_vertex])):
            weight = graph[from_vertex][to_vertex]
            if weight != 0 and distances[to_vertex] > distances[from_vertex] + weight:
                print("Graph contains a negative cycle.")
                return

    print("Shortest distances:", distances)

def initialize_distances(graph, source):
    distances = [float('inf')] * len(graph)
    distances[source] = 0
    return distances

def relax(distances, from_vertex, to_vertex, weight):
    if distances[to_vertex] > distances[from_vertex] + weight:
        distances[to_vertex] = distances[from_vertex] + weight

def dijkstra(graph, source):
    distances = initialize_distances(graph, source)
    visited = set()
    priority_queue = PriorityQueue()

    priority_queue.put((0, source))

    while not priority_queue.empty():
        current_distance, current = priority_queue.get()

        if current in visited:
            continue

        visited.add(current)

        for neighbor in range(len(graph[current])):
            weight = graph[current][neighbor]
            if weight != 0:  # Игнорируем нулевые веса (отсутствие ребра)
                if distances[neighbor] > distances[current] + weight:
                    distances[neighbor] = distances[current] + weight
                    priority_queue.put((distances[neighbor], neighbor))

    print("Shortest distances:", distances)

# Пример использования с матрицей смежности
graph_matrix = g._Graph__adjacency_matrix
# [
#     [0, 2, 4, 0, 0],
#     [0, 0, 1, 7, 0],
#     [0, 0, 0, 3, 5],
#     [0, 0, 0, 0, 2],
#     [0, 0, 0, 0, 0]
# ]

source = 13  # Начальная вершина

# Запустите алгоритм Дейкстры
# dijkstra(graph_matrix, source)

# Запустите проверку наличия отрицательных циклов с использованием Беллмана-Форда
bellman_ford(graph_matrix, source)
```

```python
import heapq
import sys

def dijkstra(matrix, start, end):
    n = len(matrix)

    if start < 0 or start >= n or end < 0 or end >= n:
        raise ValueError("Вершины не существуют в графе.")

    visited = set()
    distances = [sys.maxsize] * n
    distances[start] = 0
    previous_vertices = [None] * n

    priority_queue = [(0, start)]

    while priority_queue:
        current_distance, current_vertex = heapq.heappop(priority_queue)

        if current_vertex in visited:
            continue

        visited.add(current_vertex)

        for neighbor in range(n):
            weight = matrix[current_vertex][neighbor]

            if weight == 0:
                continue  # Пропускаем нулевые значения

            distance = current_distance + weight

            if distance < distances[neighbor]:
                distances[neighbor] = distance
                previous_vertices[neighbor] = current_vertex
                heapq.heappush(priority_queue, (distance, neighbor))

    path, current_vertex = [], end

    while previous_vertices[current_vertex] is not None:
        path.insert(0, (previous_vertices[current_vertex], current_vertex))
        current_vertex = previous_vertices[current_vertex]
    return path, distances[end]

def print_path_with_weights(path, matrix):
    for edge in path:
        start, end = edge
        weight = matrix[start][end]
        print(f"Ребро: ({start}, {end}), Вес: {weight}")

# Пример использования:
adjacency_matrix = g._Graph__adjacency_matrix




# for i in range((g._Graph__adjacency_matrix).shape[1])

i = 1
for j in range(len(g._Graph__adjacency_matrix)):
    if i == j:
        continue
    start_vertex = i
    end_vertex = j
    shortest_path, length = dijkstra(adjacency_matrix, start_vertex, end_vertex)
    if length == sys.maxsize:
        continue
    else:
        print(f"Длина маршрута для {i} с {j}: {length}")
    
    # print(f"Кратчайший маршрут для {i} с {j}:")
    # print_path_with_weights(shortest_path, adjacency_matrix)
        
        
            
        # print(shortest_path)

# shortest_path, length = dijkstra(adjacency_matrix, start_vertex, end_vertex)

# print("Кратчайший маршрут:")
# print_path_with_weights(shortest_path, adjacency_matrix)

# if length == sys.maxsize:
#     print("Пути нет")
# else:
#     print(f"Общая длина маршрута: {length}")
    
# print(shortest_path)


```
